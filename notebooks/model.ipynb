```python
import pandas as pd
```

```python
pd.set_option('display.max_columns', None)
df = pd.read_csv('../data/MachineLearningRating_v3.txt',delimiter='|')
```

```python
df.head()
```

### Data Preparation

#### Handling Missing Data

```python
(df.isnull().sum() / len(df)) * 100
```

- Drop Rows with Few Missing Values (<1%)

```python
low_missing_cols = [
    "Gender", "MaritalStatus", "mmcode", "VehicleType", "make", "Model",
    "Cylinders", "cubiccapacity", "kilowatts", "bodytype", "NumberOfDoors", 
    "VehicleIntroDate", "CapitalOutstanding"
]

df_clean = df.dropna(subset=low_missing_cols)
```

 - Impute Moderate Missing Values (1%–15%)

```python
# Fill with mode (most frequent) or "Unknown"
df_clean["AccountType"] = df_clean["AccountType"].fillna("Unknown")
df_clean["Bank"] = df_clean["Bank"].fillna("Other")
df_clean["NewVehicle"] = df_clean["NewVehicle"].fillna("Unknown")
```

- Drop Columns with High Missing Values (>60%)

```python
high_missing_cols = [
    "WrittenOff", "Rebuilt", "Converted", 
    "CustomValueEstimate", "CrossBorder", "NumberOfVehiclesInFleet"
]

df_clean = df_clean.drop(columns=high_missing_cols)
```

```python
missing_report = df_clean.isnull().mean().sort_values(ascending=False) * 100
print(missing_report[missing_report > 0])
```

#### Feature Engineering

- Convert values to DateTime

```python
df_clean['TransactionMonth'] = pd.to_datetime(df_clean['TransactionMonth'],errors='coerce')
df_clean["RegistrationYear"] = pd.to_datetime(df_clean["RegistrationYear"], errors="coerce").dt.year
df_clean['VehicleIntroDate'] = pd.to_datetime(df_clean['VehicleIntroDate'])
df_clean['CapitalOutstanding'] = pd.to_numeric(df_clean['CapitalOutstanding'], errors='coerce')
```

- Add new features

```python
df_clean["VehicleAge"] = df_clean["TransactionMonth"].dt.year - df_clean["RegistrationYear"]
df_clean["ClaimRatio"] = df_clean["TotalClaims"] / df_clean["TotalPremium"]
```

### Label Encoding

```python
top_5_ohe_cols = ["Province", "CoverType", "VehicleType", "NewVehicle", "LegalType","AlarmImmobiliser","TrackingDevice"]
df_encoded = pd.get_dummies(df_clean, columns=top_5_ohe_cols, drop_first=True)
```

```python
df_encoded.columns
```

### Modeling

```python
from sklearn.model_selection import train_test_split

df_claims = df_encoded[df_encoded["TotalClaims"] > 0].copy()
# Target
y = df_claims["TotalClaims"]
# Drop target and unrelated columns
selected_features = [
    # Binary / Categorical - Encoded
     "AlarmImmobiliser_Yes", "TrackingDevice_Yes",
   "NewVehicle_More than 6 months", "NewVehicle_Unknown",
    "LegalType_Individual", "LegalType_Partnership", "LegalType_Private company", 
    "LegalType_Public company", "LegalType_Sole proprieter",

    # Engineered & Numerical
    "CapitalOutstanding", "SumInsured", "CalculatedPremiumPerTerm", "VehicleAge",

    # Encoded Location
    "Province_Free State", "Province_Gauteng", "Province_KwaZulu-Natal",
    "Province_Limpopo", "Province_Mpumalanga", "Province_North West", 
    "Province_Northern Cape", "Province_Western Cape",

    # Encoded CoverType (selected, assuming relevant)
    "CoverType_Fire and Theft", "CoverType_Own Damage", "CoverType_Third Party", 
    "CoverType_Windscreen", "CoverType_Roadside Assistance", 
    "CoverType_Third Party Only",  # others if needed

    # Encoded VehicleType
    "VehicleType_Heavy Commercial", "VehicleType_Light Commercial", 
    "VehicleType_Medium Commercial", "VehicleType_Passenger Vehicle",
]
X_reduced = df_claims[selected_features]
# X = df_claims.drop(columns=["TotalClaims", "TotalPremium", "PolicyID", "UnderwrittenCoverID"])
X_train, X_test, y_train, y_test = train_test_split(X_reduced, y, test_size=0.2, random_state=42)
```

- Linear Regression

```python
X_train.head()
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)
```

- Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(random_state=42)
rf.fit(X_train, y_train)
```

- XGBoost Regressor

```python
from xgboost import XGBRegressor

xgb = XGBRegressor(random_state=42)
xgb.fit(X_train, y_train)
```

- Evaluate Performance

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np

def evaluate(model, name):
    y_pred = model.predict(X_test)
    print(f"\n{name} Performance:")
    print("R² Score:", r2_score(y_test, y_pred))
    print("MAE:", mean_absolute_error(y_test, y_pred))
    print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred)))

evaluate(lr, "Linear Regression")
evaluate(rf, "Random Forest")
evaluate(xgb, "XGBoost")
```
